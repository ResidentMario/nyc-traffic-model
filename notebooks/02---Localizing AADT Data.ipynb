[here](http://gis.stackexchange.com/a/212758/74038)). `fiona` will fail because of this; `QGIS GRASS` (the QGIS parser) repairs this problem and opens the file anyway.

To correct this error I do what I am learning is best practice (cf. [here](http://spatiallyadjusted.com/ogr2ogr-every-shapefile/)) and run the shapefile through `ogr2ogr`, the command line utility for `gdal`, the massive GIS cross-compiler, which repairs this error and emits a new set of files.

Nulls are associated with no data, and these are the entries which caused our initial problems:

````python
aadt.query("AADT == 'null'").head(5)
````

The geometries are in the Long Island state plane coordinate format ([EPSG:2263](http://www.spatialreference.org/ref/epsg/2263/)). We need to convert them to ordinary planar coordinates.

`geopandas` attempts to guess what coordinate reference system is, but with these gimmicky state plane systems it's often wrong (trying its best, though). We will correct it, drop the nulls (it can't reproject null coordinates, and it's not like they're useful anyway) and reproject.

````python
aadt['geometry'].isnull().astype(int).sum()
````

````python
aadt_r = aadt[aadt['geometry'].notnull()].to_crs({'init': 'epsg:4326'})
````

````python
aadt_r.head()
````

Dropping even more nulls...

````python
len(aadt_r.query('AADT == "null" or AADT_TYPE == "No data" or DATA_TYPE == "No data"'))
````

````python
aadt_rf = aadt_r.query('AADT != "null" and AADT_TYPE != "No data" and DATA_TYPE != "No data"')
````

````python
len(aadt_rf)
````

Of course, while this is data for the *state* we're interested only specifically in New York City. So we'll need to cut this data to the requisite shape.

This is easy to do, as all five boroughs appear in the top eleven totals in the `MUNI` (municipality) field.

````python
aadt_rf['MUNI'].value_counts().head(11)
````

Just for fun&mdash;this dataset contains information on almost 1500 municipalities.

````python
len(aadt_rf['MUNI'].value_counts())
````

That's useful information that could be used to e.g. scrape Wikipedia for demographic information! But not the point right now.

````python
aadt_rfg = aadt_rf[aadt_rf['MUNI'].isin(['City of Queens', 'City of Brooklyn', 'City of Bronx', 'City of Manhattan', 'City of Staten Island'])]
````

````python
len(aadt_rfg)
````
